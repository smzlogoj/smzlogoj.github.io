```r
### Aritmetica de vectores

v1 <- c(1, 2, 3)
v2 <- c(1, 1, 1)
v1
v2
```

```r
v <- v1 + v2
v
v <- v1 - 2 * v2
v
v <- 2*v1 + v2 + 1
v
v <- 2 * (v1 * v2) + 3
v
```

```r
#### Secuencias

v <- 1:30
v
v <- 30:1
v
v <- seq(-5, 5, by = 0.2)
v
v <- seq(length = 50, from = -5, by = 0.25)
v
v <- rep(c(2.5, 5.0), times = 5)
v
vc <- rep('Hola', times = 10)
vc
```

```r
###### DATA FRAMES ########
Muestra <- paste(c('m'), 1:10, sep = '_')
Muestra
Tipo <- c(rep(c('C', 'T'), 5))
Tipo
X1 <- runif(10, min = 0.1, max = 10)
X2 <- runif(10, min = 20, max = 30)
X1
X2
```

```r
df <- data.frame(muestra = Muestra, tipo = Tipo, x1 = X1, x2 = X2)
df
```

```r
summary(df)
```

```r
str(df)
```

```r
df$muestra
df$tipo
df$x1
df$x2
```

```r
df$x1 <- round(df$x1, 2)
df
df$x2 <- round(df$x2, 2)
df
```

```r
X3 <- round(runif(10, min = 50, max = 100), 2)
X3
```

```r
df$x3 <- X3
df
```
