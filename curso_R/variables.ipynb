#### - Aritmetica
R uses the usual symbols for addition +, subtraction -, multiplication *, di-
vision /, and exponentiation ^. Parentheses ( ) can be used to specify the
order of operations. R also provides %% for taking the modulus and %/% for
integer division.

```r
valor <- (1 + 1/100)^100
valor
modulo <- 17%%5
modulo
division_entera <- 17%/%5
division_entera
```

### Basic Programing

```r
# if (logical_expression) {
# expression_1
# ...
# } else {
# expression_2
# ...
# }
```

```r
numero <- 16

if (numero < 11){
    print('numero es menor de 11')
} else{
    print('numero no es menor de 11')
}
```

```r
# for (x in vector) {
# expression_1
# ...
# }
```

```r
x <- seq(1, 9, by = 2)
x

suma <- 0
for (numero in x){
    suma <- suma + numero
}
suma
```

```r
### calcular factorial
n <- 6
factorial <- 1
for (numero in 1:n){
        factorial <- factorial * numero
}
factorial
```

```r
# Bucles while
# while (logical_expression) {
# expression_1
# ...
# }
```

```r
n <- 6
factorial <- 1
contador <- 1
while (contador <= n){
    factorial <- factorial * contador
    contador <- contador + 1
}
factorial
```

```r
sumar <- function(numero1, numero2){
    suma <- numero1 + numero2
    return(suma)
}

a <- 20
b <- 32

sum <- sumar(a, b)

sum
```

```r
### Ambito de las variables
restar <- function(numero1, numero2){
    suma <- numero1 - numero2
    return(suma)
}

numero1 <- 20
numero2 <- 32

resta <- restar(a, b)

resta
```
